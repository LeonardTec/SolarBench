# ADC to Linear Actuator Position

```python
import pandas as pd
import numpy as np
import datetime
from datetime import *
```

```python
adc = list(range(379,970,1)) # range of analog digital converter (actually closer to 380, 970)
# adc = list(range(0,1025,1))
stroke = 14 # stroke of linear actuator
ret = 23.87 # length of linear actuator, retracted
ext = 37.87 # length of linear actuator, extended
# mult = stroke/(len(adc))

ADC_to_Length = lambda x : ret + stroke * (x-adc[0])/(adc[-1]-adc[0])

lengths=[]
for val in adc:
    lengths.append(ADC_to_Length(val))
    
# print(lengths)
# print(len(adc))
```

```python
len1 = 31 # length down on center column
len2 = 16 # length out on solar panel

angle_panels = lambda x : 90 - np.rad2deg(np.arccos((len1**2 + len2**2 - x**2)/(2*len1*len2)))

# print(lengths[1024]/12)
# print(angle_panels(lengths[1024]))

deg=[]
for val in lengths:
    deg.append(angle_panels(val))
    
print(deg)
```

```python
data = {'Actuator Length': lengths, 'Angle': deg, 'ADC': adc}
df = pd.DataFrame(data)

df

# should eventually (once confirmed) export df as csv to save code
```

```python
filename = 'times_positions.csv'
position_df = pd.read_csv(filename)

position_df = position_df.set_index('Unnamed: 0')
position_df
```

```python
start = datetime(year=1,month=1,day=1,hour=0,minute=0)
diff = timedelta(days=0,minutes=10,seconds=0,microseconds=0)
each = list(range(0,24*6,1))

times = []
minute = start
mo = 1
while mo < 13:
    for time in each:
        minute = datetime.strptime(str(minute),"%Y-%m-%d %H:%M:%S")
        times.append(minute.strftime("%H:%M:%S"))
        minute += diff
    mo+=1
    
# print(times)
```

```python
months = []
mo = 1
while mo < 13:
    for rep in each:
        months.append(mo)
    mo+=1

data = {'Month':months,'Time':times}
final_df = pd.DataFrame(data) 
final_df
```

```python
positions = []
ADCs = []

center = 688
low = 429    # full extension
high = 904   # full retraction

for month, time in zip(months, times):
    sunrise = position_df.loc[position_df['Month'] == month].to_numpy()[0][1]
    sunrise = datetime.strptime(sunrise,"%H:%M:%S")
    
    old_sunrise = position_df.loc[position_df['Month'] == month].to_numpy()[0][1]
    old_sunrise = datetime.strptime(old_sunrise,"%H:%M:%S")
    
    sunset = position_df.loc[position_df['Month'] == month].to_numpy()[0][5]
    sunset = datetime.strptime(sunset,"%H:%M:%S")
    
    adj = timedelta(days=0,hours=23,minutes=59,seconds=59,microseconds=0)
    time = datetime.strptime(time,"%H:%M:%S")
    
    if time < datetime.strptime('12:00:00',"%H:%M:%S"):
        time = time + adj
    if sunrise < datetime.strptime('12:00:00',"%H:%M:%S"):
        sunrise = sunrise + adj
    if sunset < datetime.strptime('12:00:00',"%H:%M:%S"):
        sunset = sunset + adj
    sun_position = (np.abs(time - sunrise)/np.abs(sunset - sunrise))*180  # sun position in degrees (0 = rise, 180 = set)
    positions.append(sun_position)
    
    if sun_position > 180:
        adc = center
    elif 0 <= sun_position <= 60:
        if time >= old_sunrise:
            adc = high
    elif 120 <= sun_position <= 180:
        adc = low
    else:
        panel_position = sun_position - 90     # panel position in degrees [-30,30]
        new_df = df.iloc[(df['Angle']-panel_position).abs().argsort()[:2]]
        adc = int(new_df.to_numpy()[0][2])
        
    ADCs.append(adc)
    
final_df['Position'] = positions
final_df['ADC'] = ADCs

final_df

# final_df.to_csv('adc_by_time.csv')

# i need to comment my code ughhh
```
